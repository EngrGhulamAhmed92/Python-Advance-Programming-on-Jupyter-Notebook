Nested Function for Names

```python
def fname(name1):
    def mname(name2):
        def lname(name3):
            return name1+" "+name2+" "+name3
        return lname
    return mname
fullname=fname("Ghulam")
lastname=fullname("Ahmed")
print(lastname("khan"))
```

Sum Multiplication Subtraction and Divison of two numbers and also round of to 3 digits

```python
def sum (num1,num2):
    return num1+num2
def sub (num1,num2):
    return num1-num2
def multi (num1,num2):
    return num1*num2
def div_no (num1,num2):
    return round(num1/num2,3) # round function is used for three decimal places
num1=float(input("Enter the first number"))
num2=float(input("Enter the second number"))
print(sum(num1,num2))
print(sub(num1,num2))
print(multi(num1,num2))
print(div_no(num1,num2))
```
